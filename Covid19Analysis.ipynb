```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#pd.set_option('display.max_columns', None)
ata_confirmed=pd.read_csv('time_series_covid19_confirmed_global.csv')
data_death=pd.read_csv('time_series_covid19_deaths_global.csv')
data_recovered=pd.read_csv('time_series_covid19_recovered_global.csv')

print(data_confirmed.shape)
print(data_death.shape)
print(data_recovered.shape)

data_confirmed.head()
```

# Un-Pivoting the data

```python
data_confirmed2 = pd.melt(data_confirmed, id_vars=['Province/State', 'Country/Region', 'Lat', 'Long'], 
                              var_name=['Date'],value_name='value')
data_death2 = pd.melt(data_death, id_vars=['Province/State', 'Country/Region', 'Lat', 'Long'], 
                           var_name=['Date'])
data_recovered2 = pd.melt(data_recovered, id_vars=['Province/State', 'Country/Region', 'Lat', 'Long'], 
                              var_name=['Date'])


print("The Shape of Confirmed is: ", data_confirmed2.shape)
print("The Shape of deaths is: ", data_death2.shape)
print("The Shape of recovered is: ", data_recovered2.shape)


data_confirmed2.head()
```

# Converting the new column to dates

```python
data_confirmed2['Date']=pd.to_datetime(data_confirmed2['Date'])
data_death2['Date']=pd.to_datetime(data_death2['Date'])
data_recovered2['Date']=pd.to_datetime(data_recovered2['Date'])
```

```python
data_confirmed2.head()
```

# Renaming column name

```python
data_confirmed2.rename(columns={'value':'confirmed'},inplace=True)
data_death2.rename(columns={'value':'death'},inplace=True)
data_recovered2.rename(columns={'value':'recovered'},inplace=True)
data_confirmed2.head()
```

# Investigating NULL values

```python
data_confirmed2.isna().sum().to_frame("# missing value")
```

# Dealing with Null values

```python
data_confirmed2.dropna(subset=['Lat','Long'],inplace=True)
data_death2.dropna(subset=['Lat','Long'],inplace=True)
data_recovered2.dropna(subset=['Lat','Long'],inplace=True)

data_confirmed2['Province/State'].fillna(data_confirmed2['Country/Region'], inplace=True)
data_death2['Province/State'].fillna(data_death2['Country/Region'], inplace=True)
data_recovered2['Province/State'].fillna(data_recovered2['Country/Region'], inplace=True)
```

```python
data_recovered2.isna().sum().to_frame("# missing values")
```

# printing shapes before the join

```python
print(data_confirmed2.shape)
print(data_recovered2.shape)
print(data_death2.shape)
```

```python
full_join = data_confirmed2.merge(data_death2[['Province/State','Country/Region','Date','death']],
                                  how="left",
                                  left_on=['Province/State','Country/Region','Date'],
                                  right_on=['Province/State','Country/Region','Date'])
full_join.head()
```

```python
full_join = full_join.merge(data_recovered2[['Province/State','Country/Region','Date','recovered']],
                                  how="left",
                                  left_on=['Province/State','Country/Region','Date'],
                                  right_on=['Province/State','Country/Region','Date'])
full_join.head()
```

# Adding Month and Year as a new Column

```python
full_join['Month-Year'] = full_join['Date'].dt.strftime('%b-%Y')
full_join.head()
```

```python
full_join.groupby(['Date']).sum()
```

```python
full_join.to_csv('CoronaVirus.csv')
```
